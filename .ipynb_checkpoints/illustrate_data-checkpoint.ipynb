# Illustrate Data

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rcParams['text.usetex'] = True
```

# M82 and M87 Upper Limits

Here we read-in and illustrate the 95% upper limits and 1/2$\sigma$ expectations resulting from the search for axion signals from M82 and M87 (See Fig. 2).

```python
M82_limits_data = np.load('data/M82_gagg_limits.npz')
M82_masses = M82_limits_data['masses']
M82_lims = M82_limits_data['lims']
M82_dn1 = M82_limits_data['dn1']
M82_expected = M82_limits_data['expected']
M82_up1 = M82_limits_data['up1']
M82_up2 = M82_limits_data['up2']


M87_limits_data = np.load('data/M87_gagg_limits.npz')
M87_masses = M87_limits_data['masses']
M87_lims = M87_limits_data['lims']
M87_dn1 = M87_limits_data['dn1']
M87_expected = M87_limits_data['expected']
M87_up1 = M87_limits_data['up1']
M87_up2 = M87_limits_data['up2']
```

```python
plt.figure(figsize=(16,9))

#this is for M82
plt.fill_between(M82_masses,M82_up1,M82_up2,color="gold",alpha=0.855,linewidth=0.)
plt.fill_between(M82_masses,M82_dn1,M82_up1,color="limegreen",alpha=0.72,linewidth=0.)
plt.plot(M82_masses,M82_expected,color="green",linestyle="dotted")
plt.plot(M82_masses,M82_lims,'k-',lw=2.)

#this is for M87
plt.fill_between(M87_masses,M87_up1,M87_up2,color="cornflowerblue",alpha=0.855,linewidth=0.)
plt.fill_between(M87_masses,M87_dn1,M87_up1,color="tomato",alpha=0.72,linewidth=0.)
plt.plot(M87_masses,M87_expected,color="firebrick",linestyle="dotted")
plt.plot(M87_masses,M87_lims,'k-',lw=2.)

plt.xlabel(r"$m_a$ [eV]", fontsize=24)
plt.ylabel(r"$g_{a\gamma \gamma}$ [GeV$^{-1}$]", fontsize=24)
plt.xscale('log')
plt.yscale('log')
plt.ylim([2e-13, 2e-10])
plt.xlim([1e-13, 1e-9])

plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

plt.show()

```

# M82 and M87 Spectra Data

Here we read-in and illustrate, for both M82 and M87, the stacked and binned OFF-subtracted NuSTAR data, our best-fit axion-induced signal model, as well as our combined best-fit power-law + axion signal model (See Fig. 3 and S4).

```python
#M82
M82_spectra_data = np.load('data/M82_data_fits.npz')
M82_Ebins = M82_spectra_data['Ebins']
M82_data = M82_spectra_data['NuSTAR_data']
M82_data_err = M82_spectra_data['NuSTAR_data_err']
M82_bf_total_signal = M82_spectra_data['bf_total_signal']
M82_bf_axion_only = M82_spectra_data['bf_axion_only']
```

```python
plt.figure(figsize=(10,8))

plt.errorbar(M82_Ebins, M82_data, yerr=M82_data_err,color = 'k', ls='none',marker = 'o', label = r"OFF-Subtracted Data")
plt.plot(M82_Ebins,M82_bf_total_signal,color='grey',marker='.',label=r"Best-Fit Power-Law + Axion Model")
plt.plot(M82_Ebins, M82_bf_axion_only, color = 'darkgoldenrod', ls=':', marker='D', markersize=4, label = r"Best-Fit Axion: $g_{a\gamma\gamma} \approx 3.7 \times 10^{-13}$ GeV$^{-1}$")


plt.text(60,2e4,r"M82",fontsize=26)
plt.xlim(30,70)
plt.yscale('log')

plt.ylim(2e-1,2e5)
plt.xlabel(r"$E$ [keV]",fontsize=24)
plt.ylabel(r"Counts",fontsize=24)
plt.legend(loc='lower left', bbox_to_anchor=(0.02, 0.03), fontsize=14, edgecolor='grey')

plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

plt.show()
```

```python
#M87
M87_spectra_data = np.load('data/M87_data_fits.npz')
M87_Ebins = M87_spectra_data['Ebins']
M87_data = M87_spectra_data['NuSTAR_data']
M87_data_err = M87_spectra_data['NuSTAR_data_err']
M87_bf_total_signal = M87_spectra_data['bf_total_signal']
M87_bf_axion_only = M87_spectra_data['bf_axion_only']
```

```python
plt.figure(figsize=(10,8))

plt.errorbar(M87_Ebins, M87_data, yerr=M87_data_err,color = 'k', ls='none',marker = 'o', label = r"OFF-Subtracted Data")
plt.plot(M82_Ebins,M87_bf_total_signal,color='grey',marker='.',label=r"Best-Fit Power-Law + Axion Model")
plt.plot(M82_Ebins, M87_bf_axion_only, color = 'darkgoldenrod', ls=':', marker='D', markersize=4, label = r"Best-Fit Axion: $g_{a\gamma\gamma} \approx 7.9 \times 10^{-13}$ GeV$^{-1}$")


plt.text(60,230,r"M87",fontsize=26)
plt.xlim(30,70)

plt.ylim(-80, 300)
plt.xlabel(r"$E$ [keV]",fontsize=24)
plt.ylabel(r"Counts",fontsize=24)
plt.legend(loc='lower left', bbox_to_anchor=(0.02, 0.03), fontsize=14, edgecolor='grey')

plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

plt.show()
```
